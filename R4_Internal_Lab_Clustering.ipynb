### 5. Calculating distortion - C distance or `inertia`

You can refer `KMeans.fit().inertia_`

Iterating values of k from 1 to 10 fit K means model
Using c distance - Get the measure for Sum of squares error.

```python
from sklearn.cluster import KMeans
```

```python
df_v_z.shape
```

```python
clusters = KMeans(1, n_init = 5)
clusters.fit(df_v_z)
```

```python
cluster_range = range(1,10)
distorsion=[]
for num_clusters in cluster_range:
    clusters = KMeans( num_clusters, n_init = 5).fit(df_v_z)
    labels = clusters.labels_
    centroids = clusters.cluster_centers_
    distorsion.append( clusters.inertia_ )
dist_error=pd.DataFrame({'clusters':cluster_range,'distorsion': distorsion})
```

```python
dist_error
```

### 6. Plotting Elbow/ Scree Plot

```python
plt.figure(figsize=(12,6))
plt.plot( dist_error.clusters, dist_error.distorsion, marker = "o" )
```

### Using optimal value of K - Cluster the data. 
Note: Since the data has more than 2 dimension we cannot visualize the data. As an alternative, we can observe the centroids and note how they are distributed across different dimensions

```python
cluster = KMeans( n_clusters = 6, random_state = 2354 )
cluster.fit(df_v_z)
```

You can use kmeans.cluster_centers_ function to pull the centroid information from the instance

```python
centroids = cluster.cluster_centers_
centroids
```

### 7. Store the centroids in a dataframe with column names from the original dataset given 

```python
centroid_df = pd.DataFrame(centroids, columns = list(df_v_z) )
centroid_df
```
