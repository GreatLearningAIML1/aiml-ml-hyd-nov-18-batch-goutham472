```python
import keras
```

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
bc=pd.read_csv("bank.csv")
bc.head()
```

```python
#dropping the columns RowNumber,customerID,surname as these are unique identifiers of the customer
bc=bc.drop(['RowNumber','CustomerId','Surname'],axis=1)
bc.head()
```

```python
#Now applying label encoding for the columsn Geography and Gender
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
bc['Geography']=le.fit_transform(bc['Geography'])
bc.head()
```

```python
bc['Gender']=le.fit_transform(bc['Gender'])
bc.head()
```

```python
#Now one hard encoding the geography
bc=pd.get_dummies(bc,columns=['Geography'])
bc.head()
```

```python
# spliting the features and target
x=bc.drop('Exited',axis=1)
y=bc['Exited']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
#Normalizing the data
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(x_train)
X_test=sc.fit_transform(x_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# create first model
model = Sequential()
#Adding 3 dense layers
model.add(Dense(12, input_dim=12,init='uniform',activation ='relu'))
model.add(Dense(8,init='uniform',activation ='relu'))
model.add(Dense(1,init='uniform',activation ='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X_train, y_train, nb_epoch=20, batch_size=10,  verbose=2)
```

```python
#predecting the results
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
```

```python
score=accuracy_score(y_test, y_pred)
score
```

```python
#Adjusting the model

# create first model
model1 = Sequential()
#Adding 3 dense layers
model1.add(Dense(6, input_dim=12,init='uniform',activation ='relu'))
model1.add(Dense(6,init='uniform',activation ='relu'))
model1.add(Dense(6,init='uniform',activation ='relu'))
model1.add(Dense(6,init='uniform',activation ='relu'))
model1.add(Dense(1,init='uniform',activation ='sigmoid'))
# Compile model
sgd = keras.optimizers.Adam(lr=0.001)
model1.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
# Fit the model
model1.fit(X_train, y_train, nb_epoch=100, batch_size=64,  verbose=2)
```

```python
y_pred1 = model1.predict(X_test)
y_pred1 = (y_pred1 > 0.5)
cm1 = confusion_matrix(y_test, y_pred1)
print(cm1)
score1=accuracy_score(y_test, y_pred1)
print(score1)
```
