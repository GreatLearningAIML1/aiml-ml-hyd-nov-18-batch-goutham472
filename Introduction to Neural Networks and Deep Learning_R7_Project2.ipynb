Loading the Data Set

```
import h5py
import numpy as np

h5f=h5py.File('/content/drive/My Drive/Colab Notebooks/SVHN_single_grey1.h5','r')

#Loading trainning and Test data

x_train=h5f['X_train'][:]
y_train=h5f['y_train'][:]
x_test=h5f['X_test'][:]
y_test=h5f['y_test'][:]
x_val = h5f['X_val'][:]
y_val = h5f['y_val'][:]

#Closing the file
h5f.close()
```

```
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
#Reshaping the data to predict in KNN
x_train1=x_train.reshape(x_train.shape[0],32*32)
x_test1=x_test.reshape(x_test.shape[0],32*32)
```

```
#Applyig KNN
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier()
knn.fit(x_train1,y_train)
```

```
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```
#Predection
y_pred = knn.predict(x_test1)
```

```
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```
#Using Neural networks

import keras
from keras.models import Sequential
from keras.layers import Dense
```

```
y_train=keras.utils.to_categorical(y_train,10)
y_test=keras.utils.to_categorical(y_test,10)
y_val=keras.utils.to_categorical(y_val,10)
```

```
x_train2=x_train.reshape(x_train.shape[0],32,32,1).astype('float32')
x_test2=x_test.reshape(x_test.shape[0],32,32,1).astype('float32')
x_val2=x_val.reshape(x_val.shape[0],32,32,1).astype('float32')
```

```
#Normalizing the data
x_train2/=255
x_test2/=255
x_val2/=255
```

```
#Building a NN Model
model=Sequential()
model.add(keras.layers.Flatten(input_shape=(32,32,1)))
model.add(Dense(200,activation ='relu'))
model.add(Dense(100,activation ='relu'))
model.add(keras.layers.Dropout(0.25))
model.add(Dense(10,activation='softmax'))
```

```
#Compile
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```
model.fit(x_train2,y_train, batch_size=64, epochs=10, 
              validation_data=(x_test2, y_test))
```

```
#Adding Batch normalizationa and decreacing learning rate

#Building a NN Model
model1=Sequential()
model1.add(keras.layers.Flatten(input_shape=(32,32,1)))
model1.add(Dense(200,activation ='relu'))
model1.add(Dense(100,activation ='relu'))
model1.add(Dense(50,activation ='relu'))
model1.add(keras.layers.normalization.BatchNormalization())
model1.add(Dense(10,activation='softmax'))
```

```
#Compile
model1.compile(loss='categorical_crossentropy',optimizer ='adam', metrics=['accuracy'])
```

```
model1.fit(x_train2,y_train, batch_size=64, epochs=10, 
              validation_data=(x_test2, y_test))
```

```
#Validation
print(model1.evaluate(x_val2,y_val))
```

```
print(model.evaluate(x_val2,y_val))
```

```
from sklearn.metrics import confusion_matrix
```

```
y_pre=model1.predict(x_val2)
```

```
y_pre1=model1.predict(x_test2)
```

```
import numpy as np

target_result=np.argmax(y_pre,axis=1)
test_result=np.argmax(y_pre1,axis=1)

```

```
#validation report
print(classification_report(y_val,target_result))
print(confusion_matrix(y_val,target_result))
```

```
#Test Report
print(classification_report(y_test,test_result))
print(confusion_matrix(y_test,test_result))
```
